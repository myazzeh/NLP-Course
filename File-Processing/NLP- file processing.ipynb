**Create a new Text File in python**

```python
open('data.txt', 'x')
```

**Open the new file for Writing task**

```python
my_file=open('data.txt', 'w')
my_file.write(' this is the first lecture\n natural language processing\n hello to my course')
# You should close the file after any writing
my_file.close()
```

**Open the new file for Reading task**

```python
f=open('data.txt', 'r')
print(f.read())
f.close()
```

**Open the new file for Reading Line by Line**

```python
f=open('data.txt', 'r')
print('sent1= '+ f.readline())
print('sent2= '+f.readline())
print('sent3= '+f.readline())
f.close()
```

```python
f=open('data.txt', 'r')

for line in f:
  print(line)
```

```python
content=['first line\n', 'second line\n', 'third line\n']
f=open('data.txt', 'w')

for line in content:
  f.write(line)

f.close()
```
